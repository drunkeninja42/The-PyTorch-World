```python
import torch
import numpy as np

torch.seed
a = torch.randint(0,3,(2,5))
print(a)
torch.einsum("ij->ji",a)
```

```python
#sum of all elements
torch.einsum("ij->",a)
```

```python
#column-wise sum

torch.einsum("ij->j",a)
```

```python
# row-wise sum
torch.einsum("ij->i",a)
```

```python
# matrix-vector multplication
v = torch.randint(3,(1,5))
torch.einsum("ij,kj->ik",a,v)
```

```python
#matrix-matrix multiplication
torch.einsum("ij,kj->ik",a,a)
```

```python
#dot product of first row with first row
torch.einsum("i,i->",a[0],a[0])
```

```python
#dot product matrix
torch.einsum("ij,ij->", a,a)
```

```python
#hadamard product ( element wise multiplication )
torch.einsum("ij,ij->ij",a,a)
```

```python
# batch-matrix multiplication
x = torch.rand((3,2,5))
y = torch.rand((3,5,3))

res = torch.einsum("ijk,ikl->ijl",x,y)
res.shape
```

```python
# finding the matrix diagonal
U = torch.rand((4,4))
print(U)
torch.einsum("ii->i",U)
```

```python
# matrix trace
torch.einsum("ii->", U)
```
